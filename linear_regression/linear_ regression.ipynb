```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
%matplotlib inline
```

```python
X = np.array(range(25))
y = X ** 1.3 + np.random.normal(10, 10, X.shape[0])
```

```python
y = y.reshape((-1, 1))
X = X.reshape((-1, 1))
```

```python
plt.plot(X, y, 'rx', 10)
plt.ylabel('Y')
plt.xlabel('X')
```

```python
def cost(X, y, theta):
    h = np.dot(X, theta)
    cos = np.sum((h - y) * ( h - y))/(2 * len(y))
    return cos
```

```python
def gradient_descent(X, y, theta, alpha, num_iters):
    m = len(y)      
    costs = [] 
    
    for _ in range(num_iters):  
        h = np.dot(X,theta)     
        theta -= alpha * (np.sum((h - y) * X, axis = 0)).reshape(theta.shape[0], 1)/m
        costs.append(cost(X, y, theta))
    return theta, costs
```

```python
def linear_regression(X, y, alpha = 0.01,num_iters = 100):
    X = np.append(np.ones((X.shape[0], 1)), X, axis = 1)
    theta = np.zeros((X.shape[1], 1), dtype = np.float64)
    theta, costs = gradient_descent(X, y, theta, alpha, num_iters)
    predicted = np.dot(X, theta)
    return predicted, theta, costs
```

```python
predicted, theta, costs = linear_regression(X, y)
```

```python
plt.plot(costs, 'b')
plt.ylabel('Cost')
plt.xlabel('No. of Interations')
```

```python
plt.plot(X, predicted, 'b')
plt.plot(X, y, 'rx', 10)
for i, x in enumerate(X):
    plt.vlines(x, min(predicted[i], y[i]), max(predicted[i], y[i]))
plt.ylabel('Y')
plt.xlabel('X')
plt.legend(('linear fit', 'data'))
plt.show()
```

```python
def feature_normaliza(X):
    mu = np.mean(X, 0)      
    sigma = np.std(X, 0)   
    def get_norm(col):
        mu = np.mean(col)      
        sigma = np.std(col)
        return (col - mu)/sigma
    return np.apply_along_axis(get_norm, 0, X), mu, sigma
```

```python
X, mu,sigma = feature_normaliza(X) 
```
